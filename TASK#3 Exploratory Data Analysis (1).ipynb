### Removing Outliers

```python
def remove_outlier(col):
    sorted(col)
    Q1,Q3=col.quantile([0.25,0.75])
    IQR=Q3-Q1
    lower_range=Q1-(1.5*IQR)
    upper_range=Q3+(1.5*IQR)
    return lower_range,upper_range
    
```

```python
lowsales,upsales=remove_outlier(df['Sales'])
df['Sales']=np.where(df['Sales']>upsales,upsales,df['Sales'])
df['Sales']=np.where(df['Sales']<lowsales,lowsales,df['Sales'])
```

```python
from sklearn.preprocessing import StandardScaler
std_scale=StandardScaler()
std_scale
```

```python
df['Postal Code']=std_scale.fit_transform(df[['Postal Code']])
df['Sales']=std_scale.fit_transform(df[['Sales']])
df['Quantity']=std_scale.fit_transform(df[['Quantity']])
df['Discount']=std_scale.fit_transform(df[['Discount']])
df['Profit']=std_scale.fit_transform(df[['Profit']])                                            
```
